```python
# Подключение библиотеки NumPy и загрузка данных

import numpy as np
data_path = "../MMO/Datasets/penguins_lab1.csv"
data = np.genfromtxt(data_path, delimiter=",")
print(data)
```

```python
# Тип переменной и форма (shape)

print ( "Data type : ", type(data) )
print ( "Data shape : ", data.shape )
print ( data[-4:] )
```

```python
# Получение типа набора данных, строки, элемента

data1 = np.genfromtxt(data_path, delimiter=",", dtype=None)
print('Shape of the dataset:', data1.shape)
print('Dataset type:', type(data1))
print('A single row of the dataset is type of:', type(data1[0]))
print('Types of elements:', type(data1[0][1]), type(data1[0][4]))
print('Dataset:')
print(data1)
```

```python
# Указание типа столбцов при загрузке данных

dt = np.dtype("U30, U30, f8, f8, f8, f8, U30")
data2 = np.genfromtxt(data_path, delimiter=",", dtype=dt)
print('Shape of the dataset:', data2.shape)
print('Dataset type:', type(data2))
print('A single row of the dataset is type of:', type(data2[0]))
print('Types of elements:', type(data2[0][1]), type(data2[0][4]))
print('Dataset slice:')
print(data2[:10])
```

```python
# Построение графиков с использованием Matplotlib

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# Данные из отдельных столбцов
bill_length_mm = [] 
bill_depth_mm = [] 
flipper_length_mm = [] 
body_mass_g = [] 

# Выполняем обход всей коллекции data2
for dot in data2:
    bill_length_mm.append(dot[2])
    bill_depth_mm.append(dot[3])
    flipper_length_mm.append(dot[4])
    body_mass_g.append(dot[5])

# Строим графики по проекциям данных
plt.figure(1)
Adelie, = plt.plot(bill_length_mm[:153], bill_depth_mm[:153], 'ro', label='Adelie')
Gentoo, = plt.plot(bill_length_mm[154:277], bill_depth_mm[154:277], 'g^', label='Gentoo')
Chinstrap, = plt.plot(bill_length_mm[278:], bill_depth_mm[278:], 'bs', label='Chinstrap')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('bill_length_mm')
plt.ylabel('bill_depth_mm')

plt.figure(2)
Adelie, = plt.plot(bill_length_mm[:153], flipper_length_mm[:153], 'ro', label='Adelie')
Gentoo, = plt.plot(bill_length_mm[154:277], flipper_length_mm[154:277], 'g^', label='Gentoo')
Chinstrap, = plt.plot(bill_length_mm[278:], flipper_length_mm[278:], 'bs', label='Chinstrap')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('bill_length_mm')
plt.ylabel('flipper_length_mm')

plt.figure(3)
Adelie, = plt.plot(bill_length_mm[:153], body_mass_g[:153], 'ro', label='Adelie')
Gentoo, = plt.plot(bill_length_mm[154:277], body_mass_g[154:277], 'g^', label='Gentoo')
Chinstrap, = plt.plot(bill_length_mm[278:], body_mass_g[278:], 'bs', label='Chinstrap')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('bill_length_mm')
plt.ylabel('body_mass_g')

plt.show()
```
